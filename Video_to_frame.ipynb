```python
import cv2
import os
import shutil
import h5py
```

```python
#Codigo para pasar de video a frames (https://medium.com/@iKhushPatel/convert-video-to-images-images-to-video-using-opencv-python-db27a128a481)
def getFrame(vidcap,video_path,sec,count):
        vidcap.set(cv2.CAP_PROP_POS_MSEC,sec*1000)
        hasFrames,image = vidcap.read()
        if hasFrames:
            cv2.imwrite(video_path[:-10]+"/frames/"+"frame"+str(count)+".jpg", image)     # save frame as JPG file
        return hasFrames

def Video_2_frame(video_path,frameRate=1/60):
    vidcap = cv2.VideoCapture(video_path)
    sec = 0
    #frameRate = 1/60 #//it will capture image in each 0.5 second
    count=1
    success = getFrame(vidcap,video_path,sec,count)
    while success:
        count = count + 1
        sec = sec + frameRate
        sec = round(sec, 2)
        success = getFrame(vidcap,video_path,sec,count)
```
